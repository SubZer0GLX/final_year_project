```python
import ForexData
import matplotlib.pyplot as plt
import tensorflow
import gpflow
import numpy as np
import pickle
import os

f = ForexData.fd()
N = 1000; cp = 'EUR_USD'; tf = 'H4'
X, Y = f.call_data(N, cp, tf)
plt.figure()
plt.plot(X, Y)
```

```python
m1 = gpflow.models.SGPR(X, Y, gpflow.kernels.RBF(1), Z=X.copy())
m1.kern.lengthscales = 0.5
m1.compile()
xx1 = np.linspace(0, 1, N).reshape(N, 1)
mean1, var1 = m1.predict_y(xx1)

plt.plot(X, mean1)
plt.plot(X,Y, 'x')
```

```python
m2 = gpflow.models.SGPR(X, mean1, gpflow.kernels.RBF(1), Z=X.copy())
m2.kern.lengthscales = 0.5
m2.kern.lengthscales.trainable = False

if os.path.isfile('./model_'+cp+'_'+tf) == True:
    with open('./model_'+cp+'_'+tf, 'rb') as fp:
        param_dict = pickle.load(fp)
        m2.assign(param_dict)

m2.compile()
gpflow.train.ScipyOptimizer().minimize(m2)

with open('./model_'+cp+'_'+tf, 'wb') as fp:
    pickle.dump(m2.read_trainables(), fp)
```

```python
xx2 = np.linspace(0, 3, 200).reshape(200, 1)
mean2, var2 = m2.predict_y(xx2)
samples2 = m2.predict_f_samples(xx2, 10)
```

```python
plt.figure(figsize=(24, 12))
plt.plot(X, Y.reshape(-1,1), 'x', mew=2)
plt.plot(xx2, mean2, 'C0', lw=2)
plt.fill_between(xx2[:,0],
                 mean2[:,0] + np.sqrt(var2[:,0]),
                 mean2[:,0] - np.sqrt(var2[:,0]),
                 color='C0', alpha=0.2)

plt.plot(xx2, samples2[:, :, 0].T, 'C0', linewidth=.33)
plt.title(cp+" "+tf)
```
